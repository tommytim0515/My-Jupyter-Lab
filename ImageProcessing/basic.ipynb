## Image Processing

It is a type of signal processing in which input is an image and output may be image or characteristics/features associated with that image. Nowadays, image processing is among rapidly growing technologies. It forms core research area within engineering and computer science disciplines too.
Image processing basically includes the following three steps:
1. Importing the image via image acquisition tools;
2. Analysing and manipulating the image;
3. Output in which result can be altered image or report that is based on image analysis.<br/>

### Sobel Edge Detection
__Edge detection__ is an image processing technique for finding the boundaries of objects within images. It works by detecting discontinuities in brightness. Edge detection is used for image segmentation and data extraction in areas such as image processing, computer vision, and machine vision.

__The Sobel operator__ performs a 2-D spatial gradient measurement on an image and so emphasizes regions of high spatial frequency that correspond to edges. Typically it is used to find the approximate absolute gradient magnitude at each point in an input grayscale image.

Linear filtering in the spacial domain uses the operation called __convolution__, which is a local operation. A convolution is basically an operation where a predefined matrix is slid across all pixels in an image, and at each pixel location, the center of the matrix is placed upon that pixel, and each pixel behind each cell of the matrix is multiplied and summed. This new summed value is placed in the exact same pixel location in the filtered image container.

The "matrix" above is called the __kernal__, figuratively the "seed" that generates the result.

As we can see from the image, when there is an edge, the slop will getting higher.
We can conclude that for a pixel, the higher slop it has, the more chance it will be on an edge.<br/>
After Sobel filtering, the magnitude of the pixel where slop is high will be larger than those where slop is low.

Let's try doing Sobel filtering for the subImg above with Sobel kernel Gx.

```python
# define the kernels here
Gx = [[1, 0, -1],
      [2, 0, -2],
      [1, 0, -1]]

Gy = [[1, 2, 1],
      [0, 0, 0],
      [-1,-2,-1]]
```

```python
# define the function for calculating the magnitude
import math
def getMagnitude(sobelFilter, img, row, col):
    result = 0
    for i in range(-1, 2):
        for j in range(-1, 2):
            if row + i < 0 or row + i >= len(img) or col + j < 0 or col + j >= len(img[0]):
                result += 0
            else:
                result += img[row + i][col + j]*sobelFilter[i + 1][j + 1]
    return abs(result)
```

```python
#define sobel filtering in x-dimension here
def sobelFilterX(img):
    result = []
    for i in range(len(img)):
        rowValue = []
        for j in range(len(img[0])):
            rowValue.append(getMagnitude(Gx, img, i, j))
        result.append(rowValue)
    return result
```

```python
result = sobelFilterX(subImg)
plt.imshow(result)
plt.show()
```

```python
# ignore the left and right boundaries and plot the magnitudes of a single row
img_set = result[5][5:96]
img_xset = np.arange(0, 91).tolist()
plt.plot(img_xset, img_set)
plt.show()
```

We can conclude that two edges have larger magnitude. <br/>
Then do the x-dimension Sobel filter for the whole picture.

```python
lena_x = sobelFilterX(lena)
plt.imshow(lena_x)
plt.title("Lena-X")
plt.show()
```

Let's then finish the part of y-dimension.

```python
#define sobel filtering in x-dimension here
def sobelFilterY(img):
    result = []
    for i in range(len(img)):
        rowValue = []
        for j in range(len(img[0])):
            rowValue.append(getMagnitude(Gy, img, i, j))
        result.append(rowValue)
    return result
```

```python
lena_y = sobelFilterY(lena)
plt.imshow(lena_y)
plt.title("Lena-Y")
plt.show()
```

The kernels can be applied separately to the image for producing separate measurements of the gradient component in each orientation. They can also be combined together to find the absolute magnitude of the gradient at each point and the orientation of that gradient. The gradient magnitude is given by:<br/>
<center>$ \left\lvert{G}\right\rvert = \sqrt{Gx^2 + Gy^2} $</center><br/>
For simplification and faster computation, we often use:
<br/>
<center> $ \left\lvert{G}\right\rvert = \left\lvert{Gx}\right\rvert + \left\lvert{Gy}\right\rvert $ </center><br/>
And then, the result will be:

```python
# define sobel filtering in two dimensions here
def sobelFilterXY(img):
    result = []
    for i in range(len(img)):
        rowValue = []
        for j in range(len(img[0])):
            rowValue.append(getMagnitude(Gx, img, i, j) + getMagnitude(Gy, img, i, j))
        result.append(rowValue)
    return result
```

```python
lena_xy = sobelFilterXY(lena)
plt.imshow(lena_xy)
plt.title("Lena-XY")
plt.show()
```
